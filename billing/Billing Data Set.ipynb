> hear below while reading only i have ignored the first 3 rows which dosent contain any usefull data

```python
df = pd.read_csv(r"Dataset (RAW).csv",low_memory=False,header=3)
df.head()
```

> in the dataframe i have observed that there was some blank columns so droped it which was unwanted 

```python
df.dropna(how='all', axis=1,inplace=True)
df.head()
```

> row 1 code explination : when they gave the data they explained that row 1 ( check above output for refference ) should be filled till the next available data with before available data

> row 2 code explination : hear i have concotinated row 0,1,2 to a single text row to 2 with saperator " ~ "

```python
df.loc[1] = df.loc[1].ffill()
df.loc[2] = df.loc[0]+" ~ "+df.loc[1]+" ~ "+df.loc[2]
df.head()
```

> row 1, 2 & 3 code explination : hear i have taken the row 2 of df ( check last df ) & converted to df column's name and assigning 0 index as data

> row 5 code explination : considered the entire df from row 3

> row 6 code explination : reseted the row index to default 

```python
new_header = df.loc[2]
new_header[0] = "date"
df.columns = new_header

df = df[3:]
df = df.rename_axis(None, axis=1)
df.head()
```

> row 1 code explination : converted the columns header to list wherex excluded the data

> row 2 code explination : hear i have performed a crosstable to list ( kid of reverse pivot table )

```python
valvar = df.columns.tolist()[1:]
df = df.melt(id_vars=["date"],value_vars=valvar)
df.head()
```

> row 1 & 2 code explination : hear to split single row column to columns in a single go converted - to ~ and all to at a go converted to multiple columns

> row 3 code explination : droped the column which was used for spliting 

> row 4 code explination : than just rearranged the columns acording to the requiremnt 

```python
df["variable"] = df["variable"].str.replace("-","~")
df[["segment","period","type","subtype"]] = df["variable"].str.split(" ~ ",expand=True)
df.drop("variable",axis=1,inplace=True)
df = df[["segment","period","subtype","value","date"]]
df
```
